**Se muestra la cantidad de Comics que tiene cada personaje**

```python
Comics_por_personaje=df_personajes[['name','comics.available']].groupby(by='name').sum()
Comics_por_personaje
```

```python
fig,ax = plt.subplots(figsize =(40, 20))
plt.pie(Comics_por_personaje['comics.available'], labels=Comics_por_personaje.index,autopct = '%0.0f%%')
plt.show()
```
